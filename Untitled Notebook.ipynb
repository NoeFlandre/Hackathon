# Preprocess & Merging the data

```python
import pandas as pd

# Load the datasets
left_accs = pd.read_csv('left_accs.csv', header=None, names=['timestamp', 'x_left', 'y_left', 'z_left'])
right_accs = pd.read_csv('right_accs.csv', header=None, names=['timestamp', 'x_right', 'y_right', 'z_right'])

# Convert Unix timestamps to datetime
left_accs['timestamp'] = pd.to_datetime(left_accs['timestamp'], unit='s')
right_accs['timestamp'] = pd.to_datetime(right_accs['timestamp'], unit='s')

# Set timestamps as the index
left_accs.set_index('timestamp', inplace=True)
right_accs.set_index('timestamp', inplace=True)

# Resample to 1 Hz (calculate mean, std, min, and max for each axis)
left_resampled = left_accs.resample('1S').agg(['mean', 'std', 'min', 'max'])
right_resampled = right_accs.resample('1S').agg(['mean', 'std', 'min', 'max'])

# Flatten multi-index column names
left_resampled.columns = ['_'.join(col) for col in left_resampled.columns]
right_resampled.columns = ['_'.join(col) for col in right_resampled.columns]

# Merge resampled left and right data on timestamp
merged_data = pd.merge(left_resampled, right_resampled, left_index=True, right_index=True, how='outer')

# Save the processed data
merged_data.to_csv('processed_acceleration_data.csv')
print("Processed data saved to 'processed_acceleration_data.csv'")
```

# Feature Creation

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute differences between left and right arm statistics
merged_data['x_mean_diff'] = merged_data['x_left_mean'] - merged_data['x_right_mean']
merged_data['y_mean_diff'] = merged_data['y_left_mean'] - merged_data['y_right_mean']
merged_data['z_mean_diff'] = merged_data['z_left_mean'] - merged_data['z_right_mean']

# Absolute differences
merged_data['x_mean_abs_diff'] = np.abs(merged_data['x_left_mean'] - merged_data['x_right_mean'])
merged_data['y_mean_abs_diff'] = np.abs(merged_data['y_left_mean'] - merged_data['y_right_mean'])
merged_data['z_mean_abs_diff'] = np.abs(merged_data['z_left_mean'] - merged_data['z_right_mean'])

# Display the first rows of the updated DataFrame
print("Updated DataFrame with new features:")
display(merged_data.head())

# Visualization: X-axis mean values and their differences
plt.figure(figsize=(12, 6))
plt.plot(merged_data.index, merged_data['x_left_mean'], label='Left X Mean', alpha=0.7)
plt.plot(merged_data.index, merged_data['x_right_mean'], label='Right X Mean', alpha=0.7)
plt.plot(merged_data.index, merged_data['x_mean_diff'], label='X Mean Difference', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Acceleration (Mean)')
plt.title('X-Axis Acceleration: Left vs Right Arm')
plt.legend()
plt.show()

# Visualization: Absolute Differences
plt.figure(figsize=(12, 6))
plt.plot(merged_data.index, merged_data['x_mean_abs_diff'], label='X Mean Abs Diff', alpha=0.7)
plt.plot(merged_data.index, merged_data['y_mean_abs_diff'], label='Y Mean Abs Diff', alpha=0.7)
plt.plot(merged_data.index, merged_data['z_mean_abs_diff'], label='Z Mean Abs Diff', alpha=0.7)
plt.xlabel('Time')
plt.ylabel('Absolute Difference')
plt.title('Absolute Differences Between Left and Right Arm Axes')
plt.legend()
plt.show()

# Save the updated DataFrame
merged_data.to_csv('processed_with_visualized_features.csv')
print("Features with differences saved to 'processed_with_visualized_features.csv'")
```
